```python
from controllers.track_recognition import *

images = [
    'images/bottom_row_no_labels.png',
    'images/ss_0.png',
    'images/ss_1.png',
    'images/ss_2.png',
    'images/ss_3.png',
    'images/ss_4.png',
    'images/ss_5.png',
    'images/ss_6.png',
    'images/ss_7.png',
]
# image = plt.imread('images/ss_5056.png')
# image = plt.imread('images/bottom_row_no_labels.png')
for image_name in images:
    image = plt.imread(image_name)
    # image = cv2.imread(image_name)
    track = extract_track(image)
    m, c = process_track_into_line(track)
    track_direction, track_offset_from_the_middle = get_track_properties(m, c, track.shape)
    print(f"Image name: {image_name}")
    print(f"m: {m:.2f}, c: {c:.2f}")
    print(f"Track direction: {track_direction:.2f}, track offset from the middle: {track_offset_from_the_middle:.2f}")

    # Display the image and the track
    track_start_coordinates, track_end_coordinates = get_track_line_coordinates(track_direction, track_offset_from_the_middle, track.shape)
    print(f"Track start coordinates: {track_start_coordinates}, track end coordinates: {track_end_coordinates}")

    plt.figure()
    plt.plot([track_start_coordinates[1], track_end_coordinates[1]], [track_start_coordinates[0], track_end_coordinates[0]], 'r-')
    plt.imshow(track, cmap='gray')
    plt.show()
```
