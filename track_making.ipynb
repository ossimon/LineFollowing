```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def paint_grass(track, grass_distance_from_track=10):
    grass = np.zeros(track.shape[:2])
    grass[:, :] = 1
    # remove grass where track is withing grass_distance_from_track pixels
    for i in range(grass_distance_from_track, track.shape[0] - grass_distance_from_track):
        for j in range(grass_distance_from_track, track.shape[1] - grass_distance_from_track):
            if np.any(track[i - grass_distance_from_track:i + grass_distance_from_track, j - grass_distance_from_track:j + grass_distance_from_track] == 0):
                grass[i, j] = 0

    return grass

track = plt.imread('textures/hard_track.png')
track_with_grass = paint_grass(track, 15)
ax, fig = plt.subplots(1, 2)
fig[0].imshow(track)
fig[1].imshow(track_with_grass, cmap='gray')
plt.show()

# save the new track with grass to a file
plt.imsave('textures/hard_track_with_grass.png', track_with_grass, cmap='gray')
```
